```python
w = 0.09
b = -3.6
```

```python
x = np.linspace(0,datos['BMI'].max(),100)
y = 1/(1+np.exp(-(w*x+b)))
```

```python
datos.plot.scatter(x='BMI',y='TenYearCHD')
plt.plot(x, y, color='black')
plt.ylim(0,datos['TenYearCHD'].max()*1.1)
plt.scatter(x, y, color='red')
plt.grid()
plt.xlabel('Indice de Masa Corporal')
plt.ylabel('Pronostico Enfermedad Coronaria 1:Positivo 0:Negativo')
plt.show()
```
